```
#The mean value of all the pixels used as threshol
total_pixels=gray_img.shape[0]*gray_img.shape[1]
sum=0
for i in range(gray_img.shape[0]):
  for j in range(gray_img.shape[1]):
    sum+=gray_img[i][j]

threshold=sum/total_pixels
```

```
threshold
```

```
threshold_image=gray_img.reshape((gray_img.shape[0],gray_img.shape[1]))
```

Using single threshold 

```
for i in range(threshold_image.shape[0]):
  for j in range(threshold_image.shape[1]):
    if threshold_image[i][j]>threshold:
      threshold_image[i][j]=1
    else:
      threshold_image[i][j]=0
```

```
plt.imshow(threshold_image,cmap='gray')
```

Using multiple thresholds

```
gray=rgb2gray(img)
```

```
for i in range(gray.shape[0]):
  for j in range(gray.shape[1]):
    if gray[i][j]>threshold:
      gray[i][j]=5
    elif gray[i][j]>0.25 and gray[i][j]<threshold:
      gray[i][j]=3
    else:
      gray[i][j]=1
```
